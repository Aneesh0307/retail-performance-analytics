```python
#Duplicate rows in the dataset
customers.duplicated().sum()
```

```python
#Dropping the Duplicate rows from the dataset
customers=customers.drop_duplicates().reset_index(drop=True)
```

```python
#After Dropping Duplicate rows in the dataset
customers.duplicated().sum()
```

```python
#Null Values in the dataset
customers.isnull().sum()
```

```python
#Filling Null Values with Mean of Age
customers=customers.fillna(customers['age'].mean())
customers.head()
```

```python
#Null values after filling with mean age
customers.isnull().sum()
```

```python
#Adding a column "Age_Group" to Segregate the customers into different age groups
def get_age_group(age):
    if age < 25:
        return '18-24'
    elif age < 35:
        return '25-34'
    elif age < 45:
        return '35-44'
    elif age < 55:
        return '45-55'
    elif age < 65:
        return '55-64'
    else:
        return '65-75'

customers['Age_Group'] = customers['age'].apply(get_age_group)
customers
```

```python
customers['age'] = pd.to_numeric(customers['age'], errors='coerce')
customers['age'] = customers['age'].round()
```

```python
#Formatting the data types of the columns in the dataset
customers['customer_id']=customers['customer_id'].astype("string")
customers['first_name']=customers['first_name'].astype("string")
customers['last_name']=customers['last_name'].astype("string")
customers['gender']=customers['gender'].astype("string")
customers['age']=customers['age'].astype("Int64")
customers['signup_date']=pd.to_datetime(customers['signup_date'])
customers['region']=customers['region'].astype("string")
customers['Age_Group']=customers['Age_Group'].astype("category")
```

```python
customers.dtypes
```

```python
#Saving the cleaned dataset as 'Customers_Cleaned.csv'
customers.to_csv('Customers_Cleaned.csv',index=False)
```

##DATA VISUALIZATIONS

####Sales Trend Over Time

```python
# Sales trend over time
sales["order_date"] = pd.to_datetime(sales["order_date"])

daily_sales = sales.groupby("order_date")["total_amount"].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales)
plt.title("Total Sales Trend Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.show()

"""This visualization shows how total sales change over time and helps identify overall growth patterns and seasonal fluctuations in sales."""
```

####Sales by Sales Channel (Online vs In-Store)

```python
# Sales comparison across sales channels
channel_sales = sales.groupby("sales_channel")["total_amount"].sum()

plt.figure(figsize=(6,4))
sns.barplot(x=channel_sales.index, y=channel_sales.values)
plt.title("Total Sales by Sales Channel")
plt.xlabel("Sales Channel")
plt.ylabel("Total Sales")
plt.show()

"""This chart compares revenue generated through online and in-store channels, helping understand which channel contributes more to total sales."""
```

####Category-wise Revenue Contribution

```python
# Revenue contribution by product category
sales_products = sales.merge(products, on="product_id", how="left")

category_sales = (
    sales_products.groupby("category")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.values, y=category_sales.index)
plt.title("Revenue by Product Category")
plt.xlabel("Total Revenue")
plt.ylabel("Category")
plt.show()

"""This visualization highlights which product categories generate the highest revenue, helping prioritize high-performing categories."""
```

####Customer Age Distribution

```python
# Distribution of customer ages
plt.figure(figsize=(8,5))
sns.histplot(customers["age"], bins=10)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

"""This distribution helps understand the age profile of customers and supports customer segmentation and targeted marketing strategies."""
```

####Top 10 Products by Revenue

```python
# Top 10 products by total revenue
product_sales = (
    sales_products.groupby("product_name")["total_amount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(8,5))
sns.barplot(x=product_sales.values, y=product_sales.index)
plt.title("Top 10 Products by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product")
plt.show()

"""This chart identifies the top revenue-generating products, which are critical for inventory planning and promotional strategies."""
```

####Return Reasons Distribution

```python
# Distribution of product return reasons
return_reason_counts = returns["return_reason"].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=return_reason_counts.index, y=return_reason_counts.values)
plt.title("Return Reasons Distribution")
plt.xlabel("Return Reason")
plt.ylabel("Number of Returns")
plt.xticks(rotation=30)
plt.show()

"""This visualization shows the most common reasons for product returns, helping identify quality or fulfillment issues."""
```

####Revenue by Store Type

```python
# Revenue comparison by store type
sales_stores = sales.merge(stores, on="store_id", how="left")

store_type_sales = sales_stores.groupby("store_type")["total_amount"].sum()

plt.figure(figsize=(6,4))
sns.barplot(x=store_type_sales.index, y=store_type_sales.values)
plt.title("Revenue by Store Type")
plt.xlabel("Store Type")
plt.ylabel("Total Revenue")
plt.show()


"""This chart compares revenue across store types and helps evaluate whether flagship or franchise stores perform better."""
```
